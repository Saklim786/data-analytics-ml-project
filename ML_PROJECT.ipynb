# DATA LOADING 

```python
# Step 1: Import libraries
import pandas as pd
import numpy as np

# Step 2: Load dataset
df = pd.read_csv("/kaggle/input/loan-data-set/loan_data_set_lyst1728376986423.csv")

# Step 3: Quick peek
print(df.shape)
df.head()
```

```python
df
```

# Missing Values Handlin

```python
# Step 4: Missing values count
df.isnull().sum().sort_values(ascending=False)
```

```python
# Step 5: Fill missing values (modern syntax—no FutureWarning)
df = df.ffill()
df = df.bfill()

# Verify again
df.isnull().sum().sort_values(ascending=False)
```

# Categorical Encodin

```python
# Step 6: Data types and unique values
print(df.dtypes)
for col in ["Gender","Married","Dependents","Education","Self_Employed","Property_Area","Loan_Status"]:
    print(col, df[col].unique())
```

# Dummy Variable Trap Avoid

```python
# Step 7: One-hot encode categorical columns
cat_cols = ["Gender","Married","Dependents","Education","Self_Employed","Property_Area","Loan_Status"]
df1 = pd.get_dummies(df, columns=cat_cols, drop_first=False)

# Check new columns
df1.filter(regex="Gender|Married|Dependents|Education|Self_Employed|Property_Area|Loan_Status").head()
```

```python
# Step 8: Drop one dummy per category (reference baseline)
to_drop = [
    "Gender_Female",
    "Married_No",
    "Education_Graduate",
    "Self_Employed_No",
    "Property_Area_Rural",
    "Loan_Status_N",
    "Dependents_3+"
]

# Drop only if column exists (safe)
to_drop = [c for c in to_drop if c in df1.columns]
df1.drop(columns=to_drop, inplace=True)

# Verify columns after drop
df1.columns.tolist()
```

# Feature & Target Selectio

```python
# Step 9: Define X (features) and y (target)
# Target: Loan_Status_Y (exists after get_dummies)
y = df1["Loan_Status_Y"]

# Features: select meaningful numeric + remaining dummies
feature_cols = [
    "ApplicantIncome","CoapplicantIncome","LoanAmount","Loan_Amount_Term","Credit_History",
    # keep all remaining dummies except target
] + [c for c in df1.columns if (
    ("Gender_" in c or "Married_" in c or "Dependents_" in c or "Education_" in c or 
     "Self_Employed_" in c or "Property_Area_" in c) and c != "Loan_Status_Y"
)]

X = df1[feature_cols]

# Quick check
print(X.shape, y.shape)
X.head()
```

# Train-Test Split

```python
# Step 10: Split data
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=0, stratify=y
)

# Check class balance
print("Train class balance:", y_train.value_counts(normalize=True).round(3).to_dict())
print("Test class balance:", y_test.value_counts(normalize=True).round(3).to_dict())
```

# Scale numeric features (Logistic Regression benefits)

```python
# Step 11: Scale numeric features (optional but recommended)
from sklearn.preprocessing import StandardScaler

num_cols = ["ApplicantIncome","CoapplicantIncome","LoanAmount","Loan_Amount_Term","Credit_History"]
scaler = StandardScaler()

x_train[num_cols] = scaler.fit_transform(x_train[num_cols])
x_test[num_cols] = scaler.transform(x_test[num_cols])

x_train.head()
```

# Train logistic regression

```python
# Step 12: Train model
from sklearn.linear_model import LogisticRegression

m1 = LogisticRegression(max_iter=1000, C=1.0, solver="lbfgs")
m1.fit(x_train, y_train)

# Coefficients peek (optional)
coef = pd.Series(m1.coef_[0], index=x_train.columns).sort_values(ascending=False)
coef.head(10)
```

# Predict and evaluate

```python
# Step 13: Predictions
y_pred = m1.predict(x_test)

# Step 14: Metrics
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

acc = accuracy_score(y_test, y_pred)
cm = confusion_matrix(y_test, y_pred)

print("Accuracy:", acc)
print("Confusion Matrix:\n", cm)
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

#  Probability threshold tuning

```python
# Step 15: Tune threshold to reduce False Positives (business control)
y_proba = m1.predict_proba(x_test)[:,1]

def evaluate_threshold(th):
    y_pred_th = (y_proba >= th).astype(int)
    cm = confusion_matrix(y_test, y_pred_th)
    acc = accuracy_score(y_test, y_pred_th)
    print(f"Threshold={th:.2f} | Accuracy={acc:.3f}")
    print(cm)

# Try few thresholds
for th in [0.3, 0.4, 0.5, 0.6]:
    evaluate_threshold(th)
```

# Cross‑validation for stability

```python
# Step 16: Cross-validation score
from sklearn.model_selection import cross_val_score

cv_scores = cross_val_score(m1, X, y, cv=5, scoring="accuracy")
print("CV Accuracy (5-fold):", cv_scores.round(3))
print("Mean:", cv_scores.mean().round(3))
```
